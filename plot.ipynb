```python
import matplotlib.pyplot as plt
import numpy as np 
import os

import matplotlib.cm as cm
from matplotlib.colors import Normalize
path = "/Users/claudiovalletta/Documents/Ph.D./Nbody"
files = []

i = 0
basicfilename = "snapshot"


for filename in np.sort(os.listdir(path)):        
        if filename.endswith(".data"):
            print(filename)
            plt.figure(figsize=(20,20))
            plt.rc('xtick', labelsize=200)
            plt.rc('ytick', labelsize=200)
            plt.tight_layout()
            data=np.loadtxt(filename)
            plt.scatter(data[:,0],data[:,1],s = 500)
            plt.scatter(data[:,2],data[:,3],s = 5000,c = 'y')
            plt.xlim([-10,10])
            plt.ylim([-10,10])
            fname = '_tmp%04d.png'%i
            plt.savefig(fname)
            files.append(fname)
            plt.close()
            print(i)
            i = i+1
            
            

os.system("opt='vbitrate=2160000:mbd=2:keyint=132:v4mv:vqmin=3:vlelim=-4:vcelim=7:lumi_mask=0.07:dark_mask=0.10:naq:vqcomp=0.7:vqblur=0.2:mpeg_quant'")

os.system("mencoder 'mf://_tmp*.png' -mf type=png:fps=10 \
          -ovc lavc -lavcopts vcodec=mpeg4:vpass=1:$opt -oac copy -o animation.mp4")

os.system("rm -rf _tmp*.png")

         






```
